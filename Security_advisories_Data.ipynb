```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

dataFrame = {
    "Title": [],
    "Date": [],
    "Severity": [],
    "Summary": [],
    "Link": []
}

dataFrame = pd.DataFrame(dataFrame)

main_url = 'https://github.com'

# URL of the GitHub Security Advisories page
for i in range (1,400):
  advisories_url = "https://github.com/advisories?page={}".format(i)

  response = requests.get(advisories_url)
  soup = BeautifulSoup(response.text, "html.parser")

  # Find all the advisory items
  advisory_items = str(soup.find_all("div", class_="d-flex position-relative"))

  soup = BeautifulSoup(str(advisory_items), 'html.parser')

  titles = []
  dates = []
  severities = []
  summaries = []
  links = []

  for div in soup.find_all('div', class_='d-flex position-relative'):
      title_elem = div.find('a', class_='Link--primary')
      title = title_elem.get_text(strip=True)
      link = title_elem['href']
      link = main_url + link

      severity_elem = div.find('span', class_='Label')
      severity = severity_elem.get_text(strip=True)

      summary_elem = div.find('div', class_='mt-1 text-small color-fg-muted')
      summary = summary_elem.get_text(strip=True)

      date_elem = div.find('relative-time')
      date = date_elem['datetime']

      titles.append(title)
      dates.append(date)
      severities.append(severity)
      summaries.append(summary)
      links.append(link)

  data = {
      "Title": titles,
      "Date": dates,
      "Severity": severities,
      "Summary": summaries,
      "Link": links
  }

  df = pd.DataFrame(data)
  dataFrame = pd.concat([dataFrame,df], ignore_index=True)
dataFrame
```
